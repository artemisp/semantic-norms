```python
import sys
sys.path.insert(1, '../../model/')
import eval
import pickle
from tqdm import tqdm
from nltk.corpus import wordnet as wn
import openai
openai.api_key = "<OPENAI-KEY>"
```

```python
DATASET = "concept_properties" # "feature_norms"
```

```python
noun2prop = pickle.load(open(f"../data/datasets/{DATASET}/noun2property/noun2prop.p", "rb"))
```

```python
candidate_adjs = []
for noun, props in noun2prop.items():
    candidate_adjs += props
candidate_adjs = list(set(candidate_adjs))
```

```python
noun2predicts = {}
for noun in tqdm(noun2prop):
    response = openai.Completion.create(
        engine="text-davinci-001",
        prompt="Use ten adjectives to describe the properties of kiwi:\n1. tart\n2. acidic\n3. sweet\n4. juicy\n5. smooth\n6. fuzzy\n7. green\n8. brown\n9. small\n10. fuzzy\nUse ten adjectives to describe the properties of {}:\n".format(noun),
        temperature=0,
        max_tokens=256,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0
    )
    props = []
    for raw_prop in dict(dict(response)['choices'][0])['text'].split('\n'):
        if len(raw_prop) >= 5 and '.' in raw_prop:
            props.append(raw_prop.split(' ')[-1].lower())
    noun2predicts[noun] = props
```

```python
eval.evaluate(noun2predicts, noun2prop)
eval.evaluate_acc(noun2predicts, noun2prop)
```

```python
## 2 shots
noun2predicts = {}
for noun in tqdm(noun2prop):
    response = openai.Completion.create(
        engine="text-davinci-001",
        prompt="Use ten adjectives to describe the properties of kiwi:\n1. tart\n2. acidic\n3. sweet\n4. juicy\n5. smooth\n6. fuzzy\n7. green\n8. brown\n9. small\n10. fuzzy\nUse ten adjectives to describe the properties of laptop:\n1. small\n2. electronic\n3. useful\n4. portable\n5. battery-powered\n6. personal\n7. metal\n8. modern\n9. new\n10. thin\nUse ten adjectives to describe the properties of {}:\n".format(noun),
        temperature=0,
        max_tokens=256,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0
    )
    props = []
    for raw_prop in dict(dict(response)['choices'][0])['text'].split('\n'):
        if len(raw_prop) >= 5 and '.' in raw_prop:
            props.append(raw_prop.split(' ')[-1].lower())
    noun2predicts[noun] = props

pickle.dump(noun2predicts, open('/nlp/data/yueyang/prototypicality/semantic-norms-main/data/datasets/concept_properties/GPT3/gpt3_predicts_2shot.p', 'wb'))
```
